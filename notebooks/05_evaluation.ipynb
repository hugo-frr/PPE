# Evaluation

```python
import numpy as np, matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,mean_absolute_error

y_test=np.load("data/processed/y_test.npy")
pred_base=np.load("data/processed/pred_base.npy")
pred_enriched=np.load("data/processed/pred_enriched.npy")
```

```python
rmse_base=mean_squared_error(y_test,pred_base,squared=False)
rmse_enriched=mean_squared_error(y_test,pred_enriched,squared=False)
mae_base=mean_absolute_error(y_test,pred_base)
mae_enriched=mean_absolute_error(y_test,pred_enriched)
rmse_base,rmse_enriched,mae_base,mae_enriched
```
