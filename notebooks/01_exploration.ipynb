# 01 - Exploration

```python
import yfinance as yf, pandas as pd, matplotlib.pyplot as plt

tickers=["AIR.PA","MC.PA","STLAM.MI"]
data=yf.download(tickers,start="2015-01-01",end="2025-12-31")["Close"].dropna()
data.head()
```

```python
plt.figure(figsize=(12,5))
for c in data.columns:
    plt.plot(data[c],label=c)
plt.legend(); plt.grid(); plt.show()
```
