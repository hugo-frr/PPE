# Enriched LSTM

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Dropout

X_train=np.load("data/processed/X_train.npy")
X_test=np.load("data/processed/X_test.npy")
y_train=np.load("data/processed/y_train.npy")
y_test=np.load("data/processed/y_test.npy")
```

```python
model_enriched=Sequential([
    LSTM(50,input_shape=(X_train.shape[1],X_train.shape[2])),
    Dropout(0.2),
    Dense(1)
])
model_enriched.compile(optimizer="adam",loss="mse")
```

```python
history=model_enriched.fit(X_train,y_train,epochs=10,batch_size=32,validation_split=0.1)
pred_enriched=model_enriched.predict(X_test)
np.save("data/processed/pred_enriched.npy",pred_enriched)
"Enriched OK" 
```
